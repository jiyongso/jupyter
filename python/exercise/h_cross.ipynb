```python
from io import StringIO
crossfile="H_cross.txt"
ff=open(crossfile, "r")
ffstr = ff.read()
ff1=ffstr.split("END\n")
crosssectiondict={}
for datastr in ff1:
    str1=StringIO(datastr)
    df=pd.read_csv(str1, delimiter="\s+")
    lc=list(df.columns)
    isotopename = "-".join(lc)
    df.columns=("energy", "crosssection")
    crosssectiondict[isotopename]=df

    
```

```python
kk = list(crosssectiondict.keys())
isoindex=4
crosssectiondict[kk[isoindex]].plot("energy", "crosssection")
plt.yscale('log')
plt.xscale('log')
plt.grid()
plt.title(kk[isoindex])
print(crosssectiondict[kk[isoindex]])
```
