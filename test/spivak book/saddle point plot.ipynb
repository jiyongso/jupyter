```python
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import matplotlib as mpl

mpl.rcParams['figure.figsize'] = (10,10)

fig = plt.figure()
ax = fig.gca(projection='3d')

X, Y = np.mgrid[-2:2:0.01, -2:2:0.01]
Z = X*X-Y*Y

y1 = np.arange(0, 1.6, 0.01)
y2 = np.arange(-1.6, 0, 0.01)
z1 = -y1*y1
z2 = -y2*y2
x1=np.zeros(y1.shape)
x2=np.zeros(y2.shape)
ax.plot_surface(X , Y , Z, cmap='autumn', cstride=2, rstride=2)
ax.plot3D(x1,y1,z1, 'b--', lw=3)
ax.plot3D(x2,y2,z2, 'b-', lw=3)

ax.w_xaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_yaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))
ax.w_zaxis.set_pane_color((1.0, 1.0, 1.0, 1.0))

ax.set_xlabel("X label")
ax.set_ylabel("Y label")
ax.set_zlabel("Z label")
ax.set_zlim(-4, 4)
ax.set_xticks([])
ax.set_yticks([])
ax.set_zticks([])
ax.set_xlabel("$x_1$", fontsize="x-large")
ax.set_ylabel("$x_2$", fontsize="x-large")
ax.set_zlabel("$x_3$", fontsize="x-large")
ax.set_axis_off()
plt.savefig("fig2_3_1.pdf")
plt.show()
```
