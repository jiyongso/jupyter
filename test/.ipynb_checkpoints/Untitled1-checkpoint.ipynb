```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

th = np.arange(0, 10.0 * np.pi, 0.01)
r1 = 2.0/(1+0.3* np.cos(1.2 *th))
r = 2.0/(1+0.3* np.cos(th))
x1, y1 = r1*np.cos(th), r1*np.sin(th)
x, y = r*np.cos(th), r*np.sin(th)
```

```python
plt.plot(x1,y1, )
```

```python
plt.plot(x,y)
```
