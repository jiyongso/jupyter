```python
class agent(object):
    
    def __init__(self, id):
        self._id = id
        self._age = None
        self._position = None
        self._speed = 4.
        self._infection_probability = 0.0             # 감염되었을 때 다른사람을 감염시킬 획률
        self._infected_period = 0
        self._cure_probability = 0.0
        self._isolated_probability = 0.2              # 감염되었을 때 isolated 될 확률
        self._infection_probability_isolated = 0.0    # isolated 되었을 때 다른 사람을 감염시킬 확률
        #self._history=[] 
        
        # time series data
        # status
        #     -1 : dead
        #     0  : not infected
        #     1  : infected but not isolated
        #     2  : infected and isolated

        self._data = pd.DataFrame({'x': pd.Series([], dtype='float'),
                   'y': pd.Series([], dtype='float'),
                   'status': pd.Series([], dtype='int')})
        
    def get_status(self) :
        if len(self._data)==0:
            return None
        else :
            return (self._data['x'].iloc[-1], self._data['y'].iloc[-1], self._data['status'].iloc[-1])
    
    def append_data(self, x, y, status=0):
        
        self._data = pd.concat([self._data, pd.DataFrame({'x': pd.Series([x], dtype='float'),
                   'y': pd.Series([y], dtype='float'),
                   'status': pd.Series([status], dtype='int')})], ignore_index=True)
        
    def set_current_infected(self):
        self._data['status'].iloc[-1]=int(1)
    
    def sec_current_cured(self):
        self._data['status'].iloc[-1]=int(0)
        
    def is_infected(self):
        return (self._data['status'].iloc[-1] >= 1)
        
    
    def calc_next_position(self, speed=None):
        if speed == None :
            speed = self._speed
        rr = np.random.random(2)
        cpos = self.get_status()
        npos = [cpos[0]+rr[0]*speed*np.cos(np.pi*2*rr[1]), 
                cpos[1]+rr[0]*speed*np.sin(np.pi*2*rr[1])]
        return (npos[0], npos[1])
        

class agentGroup(object):
    def __init__(self, Nagents=1000, Ninfected = 10, SquareWidth=50.0):
        self._Nagent = int(Nagents)
        self._infection_ratio = 0.00001
        self._death_ratio = 0.001
        self._infection_radius = 0.1
        self._isolated_radius = 0.1
        self._spatial_width = 100
        self._agents = {}
        self._agents_dead = {}
        self._lastId = 0
        self._infection_number_data = []
        self._current_infected_number = 0
        
        self._scatter_color = {-1 : "k", 0: "b", 1: "r", 2:"m"}
        
        self.init_agents(Nagents, Ninfected, SquareWidth)
        
        self.set_infection_probability(0.03)
        
        self.set_cure_probability(0.02)
        
        self.set_isolation_probability(0.2)
        
        self.set_infection_probability_isolated(0.01)
        
                 
    def init_agents(self, Nagent, Ninfected, Width):
        Nagent = int(Nagent)
        Ninfected = int(Ninfected)
        self._spatial_width = Width
        for ii in range(self._lastId+1, self._lastId+Nagent+1): 
            rp = np.random.random(2)
            x, y = rp[0]*Width-Width/2., rp[1]*Width-Width/2
            cag = agent(ii)
            cag._data = pd.DataFrame({'x': pd.Series([x], dtype='float'),
                   'y': pd.Series([y], dtype='float'),
                   'status': pd.Series([int(ii<Ninfected)], dtype='int')})
            self._agents[ii]=cag       
        self._lastId = self._lastId+Nagent
       

    def set_cure_probability(self, P):
        for ag in self._agents.values():
            ag._cure_probability = P
            
    
    def set_infection_probability(self, P=0.0001, agent_id= None):
        if agent_id == None:
            for ag in self._agents.values():
                ag._infection_probability = P
        else :
            self._agents[agent_id]._infection_probability = P
    
    
    def set_infection_probability_isolated(self, P=0.0001, agent_id = None):
        if agent_id == None:
            for ag in self._agents.values():
                ag._infection_probability_isolated = P
        else :
            self._agents[agent_id]._infection_probability_isolated = P
    
    def set_isolation_probability(self, P=0.001, agent_id = None):
        if agent_id == None:
            for ag in self._agents.values():
                ag._isolated_probability = P
        else :
            self._agents[agent_id]._isolated_probability = P
    
    def get_infected_agent_id(self):
        result = []
        for agid in self._agents.keys():
            if self._agents[agid].is_infected():
                result.append(agid)
        return result
    
    def get_infected_numbers(self):
        return len(self.get_infected_agent_id())
    
    def plot_current_position(self):
        result = pd.DataFrame({'x': pd.Series([], dtype='float'),
           'y': pd.Series([], dtype='float'),
           'status': pd.Series([], dtype='int')})
        for ag in self._agents.values():
            r1 = ag._data.iloc[-1]
            rr = pd.DataFrame({'x': pd.Series([r1[0]], dtype='float'),
           'y': pd.Series([r1[1]], dtype='float'),
           'status': pd.Series([r1[2]], dtype='int')})
            result=pd.concat([result, rr ], ignore_index=True)
        
        groups = result.groupby("status")
        ggname = {0: "normal", 1:"infected but not isolated", 2:"infected and isolated"}
        for name, group in groups:
            plt.plot(group.x, group.y, marker="o", ms=5,ls="", label = ggname[name])
            
        
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        plt.axes().set_aspect("equal")
        plt.gca().set_adjustable("box")
    
    def plot_by_step(self, step=-1):
        if step==-1:
            self.plot_current_position()
            return
        result = pd.DataFrame({'x': pd.Series([], dtype='float'),
           'y': pd.Series([], dtype='float'),
           'status': pd.Series([], dtype='int')})
        for ag in self._agents.values():
            result=pd.concat([result, ag._data.iloc[step:step+1]], ignore_index=True)
        
        groups = result.groupby("status")
        ggname = {0: "normal", 1:"infected but not isolated", 2:"infected and isolated"}
        for name, group in groups:
            plt.plot(group.x, group.y, marker="o", ms=5,ls="", c=self._scatter_color[name], label = ggname[name])
        
        plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
        plt.axes().set_aspect("equal")
        plt.gca().set_adjustable("box")    
    
    def generate_time_evolution_data(self):
        sInfect, sIsolated = None, None
        for ag in self._agents.values():
            if type(sInfect) == type(None) :
                sInfect=(ag._data.status==1).astype(int) 
                sIsolated = (ag._data.status==2).astype(int) 
                continue
            else :
                sInfect += (ag._data.status==1).astype(int) 
                sIsolated += (ag._data.status==2).astype(int) 
        self._infection_number_data = pd.DataFrame({"Ninfect" : sInfect+sIsolated, "Nisol" : sIsolated})
    
        
    def set_homogeneous_speed(self, speed= 1.):
        for ag in self._agents.values():
            ag._speed = speed
    
    
    def move(self):
        for ag in self._agents.values():
            rr = np.random.random(2)
            cpos = ag._position[-1]
            npos = [cpos[0]+rr[0]*ag._speed*np.cos(np.pi*2*rr[1]), 
                    cpos[1]+rr[0]*ag._speed*np.sin(np.pi*2*rr[1])]
            ag._position = np.vstack([ag._position, np.array(npos)])
    
    def check_move_in_box(self, x, y):
        return (np.abs(x)<self._spatial_width/2.)&(np.abs(y)<self._spatial_width/2.)
        
    def get_statistics(self):
        Ntotal = len(self._agents)
        NaccumInfected = 0 
        for ag in self._agents.values() :
            if ag._data.status.sum()>0 :
                NaccumInfected +=1
        print(Ntotal, NaccumInfected)
            

    def do_infection(self, Nloop = 100):

        next_status={}
        iids = self.get_infected_agent_id()
        Ninfected = 0
        
        # 다음 차례로 이동하면서 status는 이전 것을 그대로 복사한다.
        for ag in self._agents.values():
            nn = (0, 0)
            while(True):
                nn = ag.calc_next_position()
                if self.check_move_in_box(nn[0], nn[1]):
                    break
            current_status = ag.get_status()[-1] 
            next_status[ag._id]=[nn[0], nn[1], current_status, current_status]
        
        # infection, cured state를 update 한다.
        
        for ag1 in self._agents.values():
            pos1, status1 = ag1.get_status()[0:2], ag1.get_status()[2]
            
            for key2 in iids :
                if ag1._id == key2 :
                    continue
                
                ag2 = self._agents[key2]
                pos2, status2 = ag2.get_status()[0:2], ag2.get_status()[2]
                dist = np.sqrt((pos1[0]-pos2[0])**2+(pos1[1]-pos2[1])**2)

                if dist < self._infection_radius:
                    if (status1 == 0) and (status2 == 1) :
                        if np.random.random() < ag2._infection_probability :
                            next_status[ag1._id][3]=1

                    elif (status1==0) and (status2 ==2) :
                        if np.random.random() < ag2._infection_probability_isolated :
                            next_status[ag1._id][3]=1


            if status1==1:
                if np.random.random() < ag1._cure_probability:
                    next_status[ag1._id][3]=0

                elif np.random.random() < ag1._isolated_probability:
                    next_status[ag1._id][3]=2
                    
                    
            elif status1 == 2 :
                if np.random.random() < ag1._cure_probability:
                    next_status[ag1._id][3]=0
 
                        
        for ag in self._agents.values():        
            ag.append_data(next_status[ag._id][0], next_status[ag._id][1], next_status[ag._id][3])
            if next_status[ag._id][3] > 0 :
                Ninfected +=1
        return (Ninfected > 1)

    def loop_infection(self, Nloop=100, vervose=True, zeroquit=False):
        if Nloop < 11:
            vervose = False
        for i in range(Nloop):
            rr = self.do_infection()
            if vervose and (int(i/Nloop*10)>int((i-1)/Nloop*10)):
                print("Process : {:2.2f} %".format(i/Nloop*100))
            if (rr == False) and bool(zeroquit) :
                print("Process finised. No infected element")
                break
        if vervose:
            print("Process finished.")
        
        self.generate_time_evolution_data()
                
                
            
                
```

```python
ag = agentGroup(200, 5, 50)
ag._infection_radius = 3.4
ag.set_infection_probability(0.02)
ag.set_cure_probability(0.02)
ag.set_isolation_probability(0.02)
ag.set_infection_probability_isolated(0.005)
        
ag.loop_infection(100)
```
