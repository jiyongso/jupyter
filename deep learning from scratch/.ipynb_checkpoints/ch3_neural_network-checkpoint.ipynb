## 3 신경망

#### 3.1 페셉트론에서 신경망으로

![Neural network](./600px-Colored_neural_network.png) {width=200}

* 신경망은 입력층(Input), 은닉층(Hidden), 출력층(Output)으로 구성되어 있다. 위의 그림에서는 0층이 입력층, 1층이 은닉층, 2층이 출력층으로 **2층 신경망** 이라 한다.(가끔은 1, 2, 3층이라 이름붙이고 3층 신경망이라 부르기도 하니 주의할 것.) 2층이라 하는 것은 가중치가 입력층과 은닉층에만 부여되기 대문이다.

* 활성화 함수(activation function) :  입력신호의 총합을 출력신호로 변화하는 함수. $f=f(b+\sum w_i x_i)$ 

* 뉴런 (여기서는 노드와 동의어) : 신경망 하나하나의 처리 단위.



#### 3.2 활성화 함수

활성화 함수는 임계값을 경계로 $0$ 혹은 $1$ 의 출력값을 가지는 step function 이다. 활성화 함수를 계단 함수에서 다른 함수로 변경하는 것이 신경망의 세계로 나아가는 열쇄이다.

* 시그모이드 함수
$$
h(x) = \dfrac{1}{1+e^{-x}}
$$

* step function
$$
h(x) = \left\{ \begin{array}{ll} 1 \qquad & \text{if } x > 0 \\ 0 & \text{if } x <= 0 \end{array} \right.
$$

* ReLU 함수 
$$
f(x) = \left\{ \begin{array}{ll} x \qquad & \text{if }  x > 0 \\ 0 &\text{if } x\le 0 \end{array} \right.
$$

```python
def sigmoid(x):
    """
    Sigmoid function
    """
    return 1./(1+np.exp(-x))

def step_function(x):
    """
    Step function
    """
    return np.array(x>0, dtype = np.int)

def relu(x):
    """
    ReLU function
    """
    return np.maximum(0, x)
```

```python
xx = np.arange(-10, 10, 0.01)
yy = sigmoid(xx)
plt.plot(xx, yy, 'b-')
plt.title("Sigmoid function")
plt.grid()
```

```python
plt.plot(xx, step_function(xx))
plt.title("Step function")
plt.grid()
```

```python
plt.plot(xx, relu(xx))
plt.title("ReLU function")
plt.grid()
```

#### 3.3 다차원 배열의 계산

* Let input be $x_1,\ldots,\,x_n$ and output be $y_1,\ldots,\,y_m$. $y_i = \sum\limits_{j=1}^n w_{ij}x_j $. $X$, $Y$ 를 각각 $\{x_j\}$, $\{y_i\}$ 의 열벡터라 하고 $W=\{w_{ij}\}$ 를 $m \times n$ 행렬이라 하면 $Y=WX$ 의 행렬식으로 나타 낼 수 있다.(행렬 notation은 책마다, 저자마다 다르지만 앞으로는 일관적으로 이렇게 표현하기로 하자.)

#### 3.4 $N$-층 신경망
* $N$-층 신경망에서 입력층에서 1층으로의 신호 전달을 생각하자. $B^{(1)}$ 을 이 때의 편향이라 하고, $W^{(1)}$ 을 가중치라 하면 1층에 전달되는 가중치 $A^{(1)}=(a^{(1)}_j)$는 다음과 같이 표현 할 수 있다.
$$
A^{(1)} = W^{(1)} X + B^{(1)}
$$

* $1$-층에서의 $k$번째 활성화 함수를 $h^{(1)}_k$ 이라 하자. 이 때의 출력값을 $Z^{(1)}$ 이라 하면,
$$
Z^{(1)}_k = h^{(1)}_k(a^{(1)}_k)
$$
라 할 수 있다. 만약 모든 활성화 함수가 동일하다면 $h=h(x)$,
$$
Z^{(1)}_k = h(a^{(1)}_k)
$$
라 쓸 수 있을 것이다.

#### 3.5 출력층 설계하기

* 일반적으로 기계학습 문제는 분류(classification)과 회귀(regression) 문제로 나뉜다. 분류는 데이터가 어떤 class에 속하느냐를 결정하는 문제이며 회귀는 입력 데이터에서의 수치를 예측하는 문제이다.
* 출력층에서 사용하는 활성화함수 :  일반적으로 분류에는 소프트맥스 함수를, 회귀에는 항등함수를 사용한다. 그리고 출력층에서 사용하는 활성화 함수는 $\sigma$ 로 표현한다.
* 소프트맥스 함수(softmax function)은 다음과 같이 정의된다.
$$
y_k = \dfrac{\exp (a_k)}{\sum\limits_{i=1}^n \exp(a_i)}
$$

* 소프트맥스 함수는 지수함수를 사용하므로 $a_i>0$ 이 큰 경우 overflow 문제를 야기 할 수 있다. 이를 위해 $C$ 를 $\{a_i\}$ 값중 가장 큰 값이라 하면,

\begin{align}
y_k & = \dfrac{\exp (a_k)}{\sum\limits_{i=1}^n \exp (a_i)} = \dfrac{ \exp (a_k - C)}{\sum\limits_{i=1}^n \exp (a_i - C)}
\end{align}

이다. 즉 $a_i-C\le 0$ 이므로 overflow를 회피 할 수 있다. 

* softmax 함수의 출력값은 $[0,\,1]$ 구간에 위치한다. 따라서 확률로 해석 할 수 있다. 또한. softmax 함수는 단조증가함수이므로 $a_k$ 의 대소관계가 $y_k$ 의 대소관계에 그대로 유지된다.

* 출력층의 뉴런 수는 풀려는 문제에 맞게 적절하게 정한다. 예를 들어 분류 문제라면 최종적으로 분류되는 클래스의 수로 결정되는것이 일반적이다. 

#### 3.6 손글씨 숫자 인식

* MNIST 데이터셋을 이용한다. 

* **one-hot encoding** : 판단 후 결과값을 가장 확률이 높은 것은 1, 나머지는 0으로 저장하도록 함. 아래 load_mnist 함수의 one_hot_lable=True 이면 이렇게 하고 False 이면 그렇게 하지 않음. 

```python
# MNIST 이미지 데이터셋을 이용한 손글씨 인식.

import sys, os
import pickle 
sys.path.append(os.pardir)
from scratch.dataset.mnist import load_mnist

def get_data():
    (x_train, t_train), (x_test, t_test) = load_mnist(normalize = True, flatten = True, one_hot_label=False)
    return x_test, t_test

def init_network():
    with open("/Users/jiyong/development/jupyternotebook/deep learning from scratch/scratch/ch03/sample_weight.pkl", "rb") as f:
        network = pickle.load(f)
    
    return network

def predict(network, x):
    W1, W2, W3 = network["W1"], network["W2"], network["W3"]
    b1, b2, b3 = network["b1"], network["b2"], network["b3"]
    
    a1 = np.dot(x, W1) + b1
    z1 = sigmoid(a1)
    a2 = np.dot(z1, W2)+b2
    z2 = sigmoid(a2)
    a3 = np.dot(z2, W3)+b3
    y = softmax(a3)
    return y

def softmax(a):
    c = np.max(a)
    exp_a = np.exp(a-c)
    sum_exp_a = np.sum(exp_a)
    y = exp_a/sum_exp_a
    
    return y
```

```python
x, t = get_data()
network = init_network()

accuracy_cnt = 0

for i in range(len(x)):
    y = predict(network, x[i])
    p = np.argmax(y)
    if p == t[i]:
        accuracy_cnt += 1

print("Accuracy : "+ str(float(accuracy_cnt)/len(x)))
```
